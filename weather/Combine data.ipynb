## Combine the updated weather data with clean data

```python
import pandas as pd
import dateutil
```

```python
clean_data = pd.read_csv('./clean_datav4.csv')
weather_data = pd.read_csv('./weather_data_updated.csv')
```

```python
del clean_data['Unnamed: 0.1.1']
clean_data.head()
```

```python
weather_data.head()
```

```python
max_temp = []; mean_temp =[]; mean_wind_speed=[]; min_temp=[];total_prep=[]
for index, row in clean_data.iterrows():
    clean_data_date = dateutil.parser.parse(row['date']) ## Get the date in clean_data
    year = '{:02d}'.format(clean_data_date.year)
    month = '{:02d}'.format(clean_data_date.month)
    day = '{:02d}'.format(clean_data_date.day)
    date__ = year+'-'+month+'-'+day

    temp_df = weather_data[weather_data['validdate'].str.contains(date__)] # Find the matching date in weather data
    
    ## Transfer the data of temp, wind speed and precipitation
    max_temp.append(temp_df['max_temp'].item())
    mean_temp.append(temp_df['mean_temp'].item())
    mean_wind_speed.append(temp_df['mean_wind_speed'].item())
    min_temp.append(temp_df['min_temp'].item())
    total_prep.append(temp_df['total_prep'].item())
```

```python
clean_data['max_temp']=max_temp
clean_data['mean_temp']=mean_temp
clean_data['mean_wind_speed']=mean_wind_speed
clean_data['min_temp']=min_temp
clean_data['total_prep']=total_prep
```

```python
clean_data.to_csv('./clean_datav4.csv')
```
